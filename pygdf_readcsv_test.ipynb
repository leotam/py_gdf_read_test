```python
import pandas as pd
import pygdf
import numpy as np
```

```python
# First load in the screen metadata
data_dir = './'
bbbc021_metadata = pd.read_csv(data_dir + "BBBC021_v1_image.csv")

# now load in the mechanism of action metadata, these map to a class of compounds. compound, concentration, moa
bbbc021_moa = pd.read_csv(data_dir + "BBBC021_v1_moa.csv")

right_cols = ["compound", "concentration"]
bbbc021_merged = bbbc021_metadata.reset_index().merge(
    bbbc021_moa, how="outer", 
    left_on=["Image_Metadata_Compound", "Image_Metadata_Concentration"], 
    right_on=right_cols).set_index("index").dropna().drop_duplicates()

right_cols = ["compound", "concentration"]
bbbc021_merged = bbbc021_metadata.reset_index().merge(
    bbbc021_moa, how="outer", 
    left_on=["Image_Metadata_Compound", "Image_Metadata_Concentration"], 
    right_on=right_cols).set_index("index").dropna().drop_duplicates()
print(np.shape(bbbc021_merged))
bbbc021_merged.head()
```
